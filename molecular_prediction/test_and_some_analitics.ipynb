## Графики, статистика и простая аналитика для понимания самих данных

### Изучение размеров молекул

Исходя из структуры таблиц train и test необходимо понять как далее работать с наблюдениями.  
Например, к информации о каждом атоме(предсказание идёт на уровне атома в молекуле) можно добавить информация о самой молекуле

```python
agg_atom_count = train_data.groupby('molecule_name')\
            .agg({'id':'count'})\
            .reset_index()\
            .groupby('id')\
            .agg({'molecule_name':'count'})\
            .reset_index()
```

```python
sns.set(rc={'figure.figsize':(25,15)})
sns.barplot(x=agg_atom_count['id'], y=agg_atom_count['molecule_name'])
```

```python
sns.set(rc={'figure.figsize':(25,15)})
sns.distplot(train_data['scalar_coupling_constant'])
```

```python
sns.set(rc={'figure.figsize':(25,15)})
sns.violinplot(x = train_data['type'], y=train_data['scalar_coupling_constant'])
```

```python
sns.set(rc={'figure.figsize':(25,15)})
sns.violinplot(x = train_data['atom_index_0'], y=train_data['scalar_coupling_constant'])
```

```python
sns.set(rc={'figure.figsize':(25,15)})
sns.violinplot(x = train_data['atom_index_1'], y=train_data['scalar_coupling_constant'])
```

```python
sns.set(rc={'figure.figsize':(25,15)})
sns.violinplot(x = train_data['molecule_name'], y=train_data['scalar_coupling_constant'])
```

```python
sns.set(rc={'figure.figsize':(15,15)})
sns.heatmap(train_data.corr())
```
